4. Как найти самое часто встречаемое значение между двумя массивами? 

````python
array1 = np.array([1, 2, 3, 4, 2, 2])
array2 = np.array([2, 3, 3, 5, 1])

combined_array = np.concatenate((array1, array2))

counts = np.bincount(combined_array)

most_frequent_value = np.argmax(counts)

print(f"Самое часто встречаемое значение: {most_frequent_value}")
````

````python
import pandas as pd
````

````python
# Load dataset here
data = pd.read_csv('./student_perfomance.csv')
data
````

1. Посчитайте среднее кол-во часов уделяемое на сон студентами

````python
mean_sleep_hours = data['Sleep_Hours'].mean()
mean_sleep_hours
````

2. Сколько процентов студентов посещают внеклассовые активности?

````python
extracurricular_count = data[data['Extracurricular_Activities'] == 'Yes'].shape[0]

total_students = data.shape[0]

percentage_extracurricular = (extracurricular_count / total_students) * 100

import matplotlib.pyplot as plt

labels = ['Посещают', 'Не посещают']
sizes = [extracurricular_count, total_students - extracurricular_count]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['blue', 'red'])
plt.show()
percentage_extracurricular
````

3. Правда ли, что в среднем девушки сдают экзамены лучше, чем мужчины

````python
mean_scores_by_gender = data.groupby('Gender')['Exam_Score'].mean()
mean_scores_by_gender
````

4. Выведите 25 и 75 перцентили времени сна студентов

````python
percentile_25 = data['Sleep_Hours'].quantile(0.25)
percentile_75 = data['Sleep_Hours'].quantile(0.75)
print(percentile_25, percentile_75)
````

5. Правда ли, что те, кто спит больше(75 перцентиль и выше) сдает экзамен лучше, чем студенты с недосыпом(25 перцентиль и меньше)

````python
sleep_deprived = data[data['Sleep_Hours'] <= percentile_25]
well_rested = data[data['Sleep_Hours'] >= percentile_75]

mean_score_deprived = sleep_deprived['Exam_Score'].mean()
mean_score_rested = well_rested['Exam_Score'].mean()

print(f"Средний балл для студентов с недосыпом: {mean_score_deprived:.2f}")
print(f"Средний балл для студентов с достаточным сном: {mean_score_rested:.2f}")
````

6. Влияет ли физическая активность на результаты экзамена?

````python
import seaborn as sns
plt.figure(figsize=(6, 5))
sns.scatterplot(x='Physical_Activity', y='Exam_Score', data=data)
plt.title('Зависимость оценок на экзамене от уровня физической активности')
plt.xlabel('Уровень физической активности')
plt.ylabel('Оценка на экзамене')
plt.show()

correlation = data['Physical_Activity'].corr(data['Exam_Score'])

print(f"Коэффициент корреляции между физической активностью и оценками: {correlation:.2f}")
mean_scores_by_activity = data.groupby('Physical_Activity')['Exam_Score'].mean()

print("Средний балл на экзамене по уровням физической активности:")
print(mean_scores_by_activity)
````

7. Правда ли, что слуденты, чьи родители имеют высшее образование(College || Postgraduate), сдают экзамены лучше, чем остальные?

````python
higher_education = data[data['Parental_Education_Level'].isin(['College', 'Postgraduate'])]
others = data[~data['Parental_Education_Level'].isin(['College', 'Postgraduate'])]
mean_score_higher_education = higher_education['Exam_Score'].mean()

mean_score_others = others['Exam_Score'].mean()

print(f"Средний балл для студентов с родителями, имеющими высшее образование: {mean_score_higher_education:.2f}")
print(f"Средний балл для остальных студентов: {mean_score_others:.2f}")

if mean_score_higher_education > mean_score_others:
    print("Студенты, чьи родители имеют высшее образование, сдают экзамены лучше.")
else:
    print("Студенты, чьи родители имеют высшее образование, не сдают экзамены лучше.")
````

8. Как влияет участие родителей на результаты экзаменов

````python
involvement_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
data['Parental_Involvement'] = data['Parental_Involvement'].map(involvement_mapping)

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Parental_Involvement', y='Exam_Score', data=data)
plt.title('Зависимость оценок на экзамене от уровня участия родителей')
plt.xlabel('Уровень участия родителей')
plt.ylabel('Оценка на экзамене')
plt.show()

correlation = data['Parental_Involvement'].corr(data['Exam_Score'])

print(f"Коэффициент корреляции между участием родителей и оценками: {correlation:.2f}")

mean_scores_by_involvement = data.groupby('Parental_Involvement')['Exam_Score'].mean()

print("Средний балл на экзамене по уровням участия родителей:")
print(mean_scores_by_involvement)
````

9. Правда ли, что в семьях с доходом больше среднего дети сдают экзамены лучше?

````python
income_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
data['Family_Income'] = data['Family_Income'].map(income_mapping)

mean_income = data['Family_Income'].mean()

print(f"Средний доход семьи: {mean_income:.2f}")
high_income = data[data['Family_Income'] > mean_income]
low_income = data[data['Family_Income'] <= mean_income]
mean_score_high_income = high_income['Exam_Score'].mean()

mean_score_low_income = low_income['Exam_Score'].mean()

print(f"Средний балл для студентов из семей с доходом выше среднего: {mean_score_high_income:.2f}")
print(f"Средний балл для студентов из семей с доходом ниже или равным среднему: {mean_score_low_income:.2f}")

if mean_score_high_income > mean_score_low_income:
    print("Студенты из семей с доходом выше среднего сдают экзамены лучше.")
else:
    print("Студенты из семей с доходом выше среднего не сдают экзамены лучше.")
````

10. Посчитайте долю студентов сдавших экзамен на более чем 70 баллов в семьях с низким доходом и отсутствием доступа к интернету

````python
filtered_students = data[(data['Family_Income'] == 1) & (data['Internet_Access'] == 'No')]

if filtered_students.shape[0] > 0:
    passed_students = filtered_students[filtered_students['Exam_Score'] > 70].shape[0]

    total_students = filtered_students.shape[0]

    proportion_passed = (passed_students / total_students) * 100
else:
    proportion_passed = 0

print(f"Доля студентов, сдавших экзамен на более чем 70 баллов: {proportion_passed:.2f}%")
````
